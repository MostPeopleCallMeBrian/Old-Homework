### Filter Data Using AND & OR Operation

```python
# 26. Display the movies of the year 2008
# Boolean Condition
indexes = movies_df[movies_df['Year']!=2008].index
movies_df.drop(indexes)
```

```python
# 27. Display the movies released in 2008 and those of Genre 'Comedy' 
# Or Condition

#How can I use an OR logic to create an "AND" condition exactly?
#Gonna assume this question wants 2008 movies, then additionally comedy movies for the answer not 2008 comedies as implied
#Logic of this, get all the indexes of movies not from 2008 or not comedies, drop those leaving 2008 films and comedies
indexes = movies_df[(movies_df['Year']!=2008) & (movies_df['Genre']!='Comedy')].index
movies_df.drop(indexes)
```

```python
# 28. Display the movies released in 2011 which are of Genre 'Comedy' 
# AND Condition
#Same logic, select all movies not from 2011 and not comedies, drop them
indexes = movies_df[(movies_df['Year']!=2011) | (movies_df['Genre']!='Comedy')].index
movies_df.drop(indexes)
```
