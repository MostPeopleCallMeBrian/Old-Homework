```python
#Brian Conway
#CSCI 412 Lab 3, Fall 2021


import re
import pandas as pd #Import Pandas to work with dataframe then turn it into a csv easily

#Note that the information is grouped like so on lines:
#Name, DOB in (YYYY-MM-DD), gender(in form he/she was born X), Complete address of birth
#Complete address of current residence, longitude and latitude of current residence
#SSN
#Answer to statement 1
#Answer to statement 2
#Answer to statement 3

#Need to extract: age from date of birth
#Start with empty lists for the things we'll extract
name=[]
gender=[]
SSN=[]
age=[]
BirthAddress=[]
CurrentResidence=[]
CurrentState=[]
ResidenceLatitude=[]
ResidenceLongitude= []
S1answer=[]
S2answer=[]
S3answer =[]

#Open the file, in read mode cuz we don't need to make any changes SHORTENED DATA FOR CODING, CHANGE TO FULL SET WHEN TURNING THIS IN
f= open("raw_data.txt", "r")

#populate the lists by reading in the file and taking advantage of the pattern
for line in f:
    #Split the line into a list of words so that the tokens we need can be easily extracted
    x = line.split()
    #print(x)
    
    #Note that there is an empty line between each survey responses, this is null, so don't do anything if x is null
    #First, check if x is the first line in a survey response. If the third word is "is" then it is.   
    #We can then exctract Name, age, gender, Complete address of birth
    try:
        if(x[2]=='is'):
            patientName = (x[3]+ " "+x[4])
            name.append(patientName)
            
            #gender, if "he" is in the line, male, if not, female
            #Inverse is also true ofc not being sexist just cutting down on one check.
            if('He' in x):
                gender.append('Male')
            else:
                gender.append('Female')
            
#age, exctract the date, then (2021-the extracted year) then check and add one if he or she had a birthday already
            #regex is [0-9][0-9][0-9][0-9]-[0-9][0-9]-[0-9][0-9]
            #This will find (4 digit number)-(2 digit)-(2 digit)
            #If this looks suspiciously like YYYY-MM-DD you don't need to quit your detective job.
            pattern=re.search('[0-9][0-9][0-9][0-9]-[0-9][0-9]-[0-9][0-9]', str(x))
            #Make a string of the pattern for DOB
            pattern=pattern.group(0)
            #Make an integer of the birthYear
            birthYear=pattern[0:4]
            birthYear=int(birthYear)
            #Calculate age, current year minus birth year
            CurrentAge=2021-birthYear
            
            #Now make integers of month and day of birth
            birthMonth=int(pattern[5:7])
            birthDay=int(pattern[8:10])
            
            #Use them to check if the individual has had his or her birthday this year
            #add one of so
            #Could eventually improve this using the dateTime module but correct as of current day
            if(birthMonth<=10 & birthDay<=6):
                CurrentAge=CurrentAge+1

            age.append(CurrentAge)
        #if(pattern in str(x)):
        #    age.append(pattern.group(0))
        if(x[0]=='SSN'):
            SSNpattern=re.search('[0-9][0-9][0-9]-[0-9][0-9]-[0-9][0-9][0-9][0-9]', str(x))
            SSN.append(SSNpattern.group(0))


        if(x[0]=='S1,'):
            if(x[len(x)-2]=='STRONGLY'):
                S1answer.append(x[len(x)-2] + " " + x[len(x)-1])

            else:
                S1answer.append(x[len(x)-1])
       
        if(x[0]=='S2,'):
            if(x[len(x)-2]=='STRONGLY'):
                S2answer.append(x[len(x)-2] + " " + x[len(x)-1])

            else:
                S2answer.append(x[len(x)-1])

        if(x[0]=='S3,'):
            if(x[len(x)-2]=='STRONGLY'):
                S3answer.append(x[len(x)-2] + " " + x[len(x)-1])

            else:
                S3answer.append(x[len(x)-1])           
#Except block to prevent program from crashing when fed an empty line.
#Nothing to do on an empty line so just pass
    except:
          pass  

#zipped = list(zip(name,gender,SSN,age,BirthAddress,CurrentResidence,CurrentState,ResidenceLatitude,ResidenceLongitude,S1answer,S2answer,S3answer))
#df = pd.DataFrame(zipped, columns=['Name','Gender','SSN','Age','Birth Address','Current Residence','State','Lat','Long','S1','S2','S3'])

```
