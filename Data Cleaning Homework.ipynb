```python
import numpy as np
import pandas as pd
import csv
import xml.etree.ElementTree as ET


#A method for finding the zip codes
def IsZip(elem):
    if (elem.tag=='tag') and (elem.attrib['k'] == "addr:postcode"):
        return True

#List of Zips
Zips = []

#Parsing the OSM and adding all the zips to the list
tree = ET.parse('/Milwaukee.osm')
ZipCount = 0
for elem in tree.iter():
    if (IsZip(elem) == True):
        Zips.append(elem.attrib['v'])

#Method for formatting the zips. Try to return the zip as an int and fail if it's not a zip such as if a word is entered as the zip.
def StringZip(string):
    trim=string[0:5]
    try:
        return int(trim)
    except:
        return 'Error, needs fixed manually'

#Adding all the fixed zips to a new list. 
FixedZips =[]
for a in Zips:
    FixedZips.append(StringZip(a))
    
#Outputting the fixed zips.
print(FixedZips)
        

```
