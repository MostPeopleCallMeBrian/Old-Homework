# 2. Dealing with imbalances 

We have three classes (happiness, satisfaction, trust), each taking three values. Classifiers would be built for each class separately. That is, given a class such as happiness, you would build a classifier that identifies its outcome (happy? unhappy? neutral?) given the features. The problem is that there is some imbalance because 'neutral' is less prevalent. Normally, we'd deal with imbalances by a mix of under- and over-sampling (i.e. generating synthetic data). But there are other approaches out there. This section completes our class discussion by showing you two such approaches.

2.1 Careful design of sub-samples for bootstrap aggregating (i.e. bagging)
There are many ways to dealing with imbalances. In the class, we mostly discuss under- and over-sampling (by creating synthetic data). In this section, we complement the class discussion by looking at Bootstrap aggregating), also commonly known as bagging. In short, a decision tree is very sensitive to the specific data on which you train it which means that it can easily overfit. Bagging will instead divide the data into many random sub-samples, build a decision tree on each one, and combine the predictions made by each tree. It's a good approach to avoid over-fitting, not an approach to deal with imbalance. But we'll see in that it can be extended to deal with both over-fitting and class imbalance.

Because you end up with several trees, it's an ensemble method, i.e. the prediction is given by looking at the consensus among many classifiers. Note that bagging is not limited to trees. You could use it on any machine learning algorithm that has an easy tendency to overfit.

The following code uses scikit-learn's built-in bagging process. It's just an example of how bagging works. You can run it, because these variables should exist in your data.



```python
import pandas
from sklearn import model_selection
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
seed=7
kfold = model_selection.KFold(n_splits=10)# random_state=seed)
# list the predictive attributes. This example just takes 4 but you have more!
featureNames=[]
for i in range(20,70):
    featureNames.append('cp17i0'+str(i))
X = data[featureNames]#separate the attributes as X...
Y = data[['Satisfaction']] #pick the class you want to predict

# we make a 'bag of classifiers'. We specify that each classifier is a tree and we use 20 trees.
model = BaggingClassifier(base_estimator=DecisionTreeClassifier(), n_estimators=10, random_state=seed)
results = model_selection.cross_val_score(model, X, Y.values.ravel(), cv=kfold)
print(results.mean())
```
