# 3. Cleaning

At this stage, you have a file that is properly formatted as a CSV so it can be opened as a Pandas DataFrame. We have some sense of what's in it. It's now time to clean the content, for at least two reasons:
* It's not because answers are nicely formatted as columns that they _make sense_. The objective will be to identify anomalies and drop records that express them. <i>(There are many ways to deal with anomalies, and 'drop the whole patient' is the simplest but not the only one.)</i>
* We may choose to filter the data because we focus on some specific parts of it.

````python
#Q3.1.> We will focus on women who live in [Southern states](https://en.wikipedia.org/wiki/Southern_United_States). 
#Eliminate everyone else. 
#Southern states are define as:
#Alabama - AL
#Arkansas - AK
#Delaware - DE
#Florida - FL
#Georgia - GA
#Kentucky - KY
#Louisiana - LA
#Maryland - MD
#Mississippi - MS
#North Carolina - NC
#Oklahoma - OK
#South Carolina - SC
#Tennessee - TN
#Texas - TX
#Virginia - VA
#West Virginia - WV

#List the abbreviaations cuz that's how they are in the data
SouthStates = ['AL','AK','DE','FL','GA','KY','LA','MD','MS','NC','OK','SC','TN','TX','VA','WV']

#Remove male patients from the data
df = df[df.gender!='Male']
#Remove remaining patients not in southern states from the data
df = df[df.state.isin(SouthStates)]
#Rest of the dataframe is women in southern states
df


````

````python
#Q3.2 Using Pandas, identify anomalies in the data in terms of answers to statements S1, S2, and S3. 
#The question states what scale must be used, but patients may have ignored instructions nonetheless.
#If you identify a record/patient with anomalies, remove it from the dataset. 
#That means that patients who do not provide answers to S1-S2-S3 in the right scale are removed.

#Pretty simple fix, just three lines on S1 S2 AND S3 similar to the one that trimmed down to patients in southern states
#A similar list will help too, of valid answers to the survey
validAnswers=['STRONGLY DISAGREE','DISAGREE','NEUTRAL','AGREE','STRONGLY AGREE']

#Now trim, applying to each question
df= df[df.S1.isin(validAnswers)]
df= df[df.S2.isin(validAnswers)]
df= df[df.S3.isin(validAnswers)]

df
#your code here
````

> Replace the values of S1. When the answers are either "AGREE" or "STRONGLY AGREE", write 'True'; otherwise write 'False'. In other words, you'll dichotomize S1 into 'True' (happy) and 'False' (unhappy).
>

````python
#your code here
#list happy little answers here, call me Bob Ross
happyLittleAnswers = ['STRONGLY AGREE','AGREE']
#Unhappy answers
unhappyAnswers = ['STRONGLY DISAGREE','DISAGREE','NEUTRAL']

#Syntax:
#DataFrame.loc[condition, column_name] = new_value

#Now apply the syntax for the happy and unhappy answers
df.loc[df.S1.isin(happyLittleAnswers), 'S1'] = True
df.loc[df.S1.isin(unhappyAnswers), 'S1'] = False

df
````

> Q4.2
> This is the end. Everything is in place, hence you can run a machine learning algorithm. Randomly split the data into a training set (80%) and a testing set (20%). Then, predict whether a person is happy or unhappy based on all other features. 

````python
#Q4.2.1. Extract needed columns S1,S2, S3 from the data frame
#one line of code, just make the DF only s1 2 and 3
df = df[["S1","S2","S3"]]
df 
````

````python
#Q4.2.2. import the needed libraries for classfication
#your code here
import numpy as np #Not sure if I need this or not but can't hurt
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.preprocessing import OrdinalEncoder
````

````python
#Q4.2.3. split the data into train and test sets

#Y is the target, make it a boolean series so that the dTree can use it
y = df['S1'].astype('bool')


#x is all the other so just drop the Y values out of it
x = df.drop(['S1'], axis = 1)

#Test train split, this is a pretty famously common line of code
#Setting to 0.2 test size arbitrarily cuz not specified otherwise
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
````

````python
#Find the unique values
x_train['S2'].unique()
````

````python
#Set up an ordering for the values
s2_ans_ordering = ['STRONGLY DISAGREE', 'DISAGREE', 'AGREE', 'STRONGLY AGREE']
s3_ans_ordering = ['STRONGLY DISAGREE', 'DISAGREE', 'AGREE', 'STRONGLY AGREE']
````

````python
#creaing the ordinal encoder for our data
ordi = OrdinalEncoder(categories=[s2_ans_ordering, s3_ans_ordering])
````

````python
#we need to fit before doing the following
ordi.fit(x_train[['S2','S3']])
````

````python
#test is basically x_train from my understading
new_train_x = ordi.transform(x_train[['S2','S3']])
````

````python
#Q4.2.4. construct the decision tree classifier model (criterion is entropy and depth is 4) and train it (using fit())

#naming it dTree so it works with the prof's code. setting depth to 4. Entropy is default criterion so no need to specify.
dtree = tree.DecisionTreeClassifier(max_depth=4)
#Make the dtree fit training data
dtree.fit(new_train_x, y_train)
````

````python
#And finally, this visualizes the tree in textual form
from sklearn.tree.export import export_text #THIS requires at least version 0.21.2 of scikit-learn

resultingTree = export_text(dtree,feature_names=list(x.columns))
print(resultingTree)
````

````python
#Q4.2.5. predict the results for the new data x_test

#Need to do the same ordinal fitting with the test as training data
ordi.fit(x_test[['S2','S3']])
x_test = ordi.transform(x_test[['S2','S3']])
y_pred=dtree.predict(x_test)
````

````python
#Q4.2.6. Display the prediction accuracy score using sklearn.metrics
from sklearn import metrics
#One line of code for this
print(metrics.accuracy_score(y_test, y_pred))
````

````python
#This loads the cleaned CSV file
import pandas as pd
data = pd.read_csv("raw_data.csv") #depending on where you've put it/named it, change this line

#-----HERE----- you can call your function to dichotomize S1 into True/False
#If you run this script as it is, it won't do a binary classification because S1 isn't binary. But you'll still get a tree.

#This divides the data into features vs the class
from pandas import DataFrame
Y = DataFrame(data,columns=['S1'])
X = pd.get_dummies(data[['gender','age','state']],drop_first=True) #get_dummies helps handle categorical data in this case

#This runs a decision tree on the data
from sklearn import tree
decisionTree = tree.DecisionTreeClassifier(max_depth=2)
decisionTree = decisionTree.fit(X, Y)

#And finally, this visualizes the tree in textual form
from sklearn.tree.export import export_text #THIS requires at least version 0.21.2 of scikit-learn
resultingTree = export_text(decisionTree,feature_names=list(X.columns))
print(resultingTree)
````
